### 2.2 Understanding the Missingness


#### 2.2.1 Explanation of Missing Values

```python
missing_df.head(20)
```

- The columns PoolQC, MiscFeature, Alley, Fence, and FireplaceQu do not present incomplete data.
- So, these missing values indicate the absense of the related feature
- Ex:
    - PoolQC is missing when the house does not have a pool
    - Alley is missing when the house does not have alley access
    - Fence is missing because there is no fence for the corresponding house

- Therefore, these are not true missing values but meaningful indicators of absence.


#### 2.2.2 LotFrontage and Neighborhood

```python
lotfrontage_by_neighborhood = df.groupby('Neighborhood')['LotFrontage'].median()

lotfrontage_by_neighborhood
```

##### Explanation

- The median LotFrontage varies significantly across different neighborhoods.
- Neighborhoods like ClearCr have a higher median frontage, while others like BrDale and MeadowV have much smaller values.
- This indicates that LotFrontage is highly dependent on the Neighborhood.
- Therefore, the missing values in LotFrontage are not random.


#### 2.2.3 Missingness Mechanism

##### Missingness Mechanism Classification

- LotFrontage → MAR (Missing at Random)
  - The missing values in LotFrontage depend on the Neighborhood.
  - Since LotFrontage varies across neighborhoods, the missingness is related to another observed variable (Neighborhood), making it MAR.

- PoolQC → MNAR (Missing Not at Random)
  - PoolQC is missing when a house does not have a pool.
  - The missingness is directly related to the absence of the feature itself, not random, making it MNAR.

- Electrical → MCAR (Missing Completely at Random)
  - The missing values in Electrical appear to be very few and do not show any clear pattern or dependency on other variables.
  - Therefore, it can be considered as missing completely at random.

### 2.3 Imputation Strategies

#### 2.3.1 Fill missing values with None

```python
cols = ['PoolQC', 'Alley', 'Fence', 'MiscFeature', 'FireplaceQu']
df[cols] = df[cols].fillna("None")
df[cols].head()
```

```python
df[cols].isnull().sum()
```

#### 2.3.2 Impute LotFrontage using the median per Neighborhood

```python
print("Missing before: ", df['LotFrontage'].isnull().sum())
df['LotFrontage'] = df['LotFrontage'].fillna(
    df.groupby('Neighborhood')['LotFrontage'].transform('median')
)
print("Missing after:", df['LotFrontage'].isnull().sum())
```

#### 2.3.3 Impute GarageYrBlt

```python
df[['GarageType', 'GarageYrBlt']].head(50)
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
```

```python
df[['GarageType', 'GarageYrBlt']].head(50)
```

##### Explanation

- The missing values in GarageYrBlt occur when GarageType is also missing.
- That means the house property does not have a garage.
- Therefore, these are not true missing values but represent the absence of a garage.
- Imputing with mean or median would introduce incorrect information by assigning a construction year to a non-existent garage.
- So, missing values were filled with 0 to clearly indicate that no garage exists.


#### 2.3.4 Fill Missing Values using Sklearn

```python
from sklearn.impute import SimpleImputer
col = ['MasVnrArea']
imputer = SimpleImputer(strategy='mean')
print("Missing before: ", df['MasVnrArea'].isnull().sum())
df[col] = imputer.fit_transform(df[col])
print("Missing after: ", df['MasVnrArea'].isnull().sum())
```

#### 2.3.5 Impute All

```python
total_missing = df.isnull().sum().sum()
print("Total Missing: ", total_missing)
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df[['MasVnrType', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
    'Electrical', 'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond']]
```

```python
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
cols_none = [
    'MasVnrType',
    'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
    'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'
]

df[cols_none] = df[cols_none].fillna('None')
```

```python
print("Total missing:", df.isnull().sum().sum())
```

##### Explanation

- Columns related to basement, garage, and masonry veneer were filled with "None" because missing values indicate the absence of these features rather than missing data.
- The Electrical column had only one missing value and was imputed using the mode, as it is a categorical variable with minimal missingness
